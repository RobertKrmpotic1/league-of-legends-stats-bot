```python
import numpy as np
import altair as alt
import pandas as pd
import selenium
df = pd.DataFrame(data = np.arange(0,300, step=5),columns=["Armor"])
df["% dmg taken"] = 100 / (100 +df["Armor"] )
df["% dmg blocked"] = 1- df["% dmg taken"]
df.head()
```

```python
chart  = alt.Chart(df).mark_line().encode(
    x='Armor',
    y=alt.Y('% dmg blocked', axis=alt.Axis(format='%')),
    
).properties(
    title = "% dmg blocked depending on Armor",
    width=800,
    height=400
)
chart
```

```python
df["% dmg blocked increase"]  =    df["% dmg blocked"].shift(-3)-df["% dmg blocked"]
alt.Chart(df.iloc[4:,:]).mark_line().encode(
    x=alt.X('Armor', scale=alt.Scale(domain=(20,280))),
    y=alt.Y('% dmg blocked increase', axis=alt.Axis(format='%')),
    
).properties(
    title = "% dmg blocked increase with buying Cloth Armor (15 Armor, 300 gold)",
    width=800,
    height=400
)
```

```python
df_hp = pd.DataFrame(data = np.arange(550,6000, step=50),columns=["HP"])
df_hp["% HP Increase"] =  1-df_hp["HP"] /  (df_hp["HP"] + 150)

alt.Chart(df_hp).mark_line().encode(
    x=alt.X("HP"),
    y=alt.Y('% HP Increase', axis=alt.Axis(format='%')),
    
).properties(
    title = "% HP increased with buying Ruby Crystal (400 gold 150 HP)",
    width=800,
    height=400
)
```
